```python
import sys
sys.path.insert(1, '../functions')
import fractals
import matplotlib.pyplot as plt
```

```python
# Barnsley fern
x_coords, y_coords = fractals.BarnsleyFern()

plt.figure(figsize=(3,5), dpi=1000)
plt.scatter(x_coords, y_coords, s=0.2, color='k', marker='.', facecolor='k', edgecolor='none')
plt.axis('off')
plt.tight_layout()
plt.savefig('../images/barnsley_fern.png');
```

```python
# Serpienski triangle
rows = fractals.SierpinskiTriangle()
plt.figure()#figsize=(1,2))
plt.matshow(rows, cmap=plt.cm.binary)
plt.axis('off')
plt.savefig('../images/serpienski_triangle.png', dpi=400)
plt.show();
```
