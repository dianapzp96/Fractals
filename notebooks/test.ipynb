```python
import sys
sys.path.insert(1, '../functions')
import fractals
import matplotlib.pyplot as plt
```

```python
# Barnsley fern
x_coords, y_coords = fractals.BarnsleyFern()

plt.figure(figsize=(5,8))
plt.scatter(x_coords, y_coords, s=0.3, color='k', marker='.')
plt.axis('off')
plt.savefig('../images/fern.png', dpi=400)
plt.show();
```

```python
# Serpienski triangle
rows = fractals.SierpinskiTriangle()
plt.figure(figsize=(5,8))
plt.matshow(rows, cmap=plt.cm.binary)
plt.axis('off')
plt.savefig('../images/triangle.png', dpi=400)
plt.show();
```
