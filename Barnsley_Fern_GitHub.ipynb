```python
import matplotlib.pyplot as plt
```

```python
# Barnsley fern
x_coords, y_coords = BarnsleyFern()

plt.figure(figsize=(5,8))
plt.scatter(x_coords, y_coords, s=0.3, color='k', marker='.')
plt.axis('off')
plt.savefig('results/fern.png', dpi=400)
plt.show();
```

```python
# Serpienski triangle
rows = SierpinskiTriangle()
plt.figure(figsize=(5,8))
plt.matshow(rows, cmap=plt.cm.binary)
plt.axis('off')
plt.savefig('results/triangle.png', dpi=400)
plt.show();
```
